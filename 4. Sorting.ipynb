```python
def bubbleSort(seq):
    length = len(seq)
    for i in range(length - 1):            # 0부터 length - 1 까지
        for j in range(length - i - 1):    # j번째와 j+1번째 요소와 비교해서 스왑합니다.
            if seq[j] > seq[j + 1]:
                seq[j], seq[j + 1] = seq[j + 1], seq[j]
        print("{}번째 : {}".format((i+1), seq))
                
    return seq

if __name__ == "__main__":
    seq = [4, 5, 2, 1, 6, 2, 7, 10, 13, 8]
    print("정렬 전 : {}".format(seq))
    seq = bubbleSort(seq)
    
    print("정렬 후 : {}".format(seq))
```

```python
# 선택정렬
# 리스트의 해당 순서에 어떤 원소를 넣을지 선택하며 정렬하는 방식입니다.
# 내림차순으로 정렬한다면,
# 첫번째 원소에는 가장 작은값을 찾아 넣어줄거기 때문에
# 리스트에서 가장 작은값을 찾아 첫번째 요소와 위치를 바꿉니다.
# 두번째는 남은 요소들중 가장 작은 값을 찾아 두번째 요소와 위치를 바꿉니다.
# 반복하여 내림차순으로 정렬을 완료합니다.
# list_a = [11,3,28,43,9,4]

def selection_sort(seq):
    length = len(seq)
    
    for i in range(length - 1):
        min_i = i
        for j in range(i + 1, length):
            if seq[min_i] > seq[j]:
                min_i = j
        seq[i], seq[min_i] = seq[min_i], seq[i]
        print("{}번째 : {}".format(i+1, seq))
        
    return seq

if __name__ == "__main__":
    list_a = [11,3,28,43,9,4]
    
    print(selection_sort(list_a))
```

```python
# 삽입정렬(Insertion sort)은 자료 배열의 모든 요소를 앞에서부터 차례대로 이미 정렬된 배열 부분과 비교하여, 
# 자신의 위치를 찾아 삽입하는 방법입니다.
def insertion_sort(li):
    length = len(li)
    
    
    for i in range(1, length):
                                    # j는 1번째 부터 시작하여 그 앞의 값들과 비교하여 자신의 자리를 찾습니다.
        for j in range(i, 0, -1):   # 자신보다 큰 값들과 자리를 바꾸며 자신보다 작은 값 뒤까지 반복합니다.
            if li[j] < li[j - 1]:
                li[j], li[j - 1] = li[j - 1], li[j]
                
        print("{}번째 : {}".format(i - 1,li))
    return li
        
li = [5, 4, 3, 7, 6, 2]
insertion_sort(li)
```

```python
# 놈 정렬
# 앞으로 이동해서 잘못 정렬된 값을 찾은 후 올바른 위치로 값을 교환하며 뒤로 이동하는 방법입니다.

def gnome_sort(li):
    
    i = 0
    
    while i < len(li):           # 인덱스가 0이거나 현재 인덱스의 요소가 전 요소보다 크다면 다음 인덱스로 이동
        if i == 0 or li[i - 1] <= li[i]:
            i += 1
            
        else:                    # 현재 인덱스의 요소가 전 요소보다 작다면 swap후에 전 인덱스로 이동
            li[i], li[i - 1] = li[i - 1], li[i]
            i -= 1
            
        print(li)
        
gnome_sort([5, 3, 2, 4])
```

```python
# 카운트 정렬
# 작은 범위의 정수를 정렬할 때 사용합니다 -> 정수의 범위가 크면 매우 비효율적
# 숫자 발생 횟수를 계산하는 누적 카운트를 사용하며
# 누적 카운트를 갱신해서 순서대로 숫자를 직접 배치하는 방식

# collections.defaultdict 모듈에서 제공하는 추가적인 딕셔너리 타입을 활용
# > 기본값이 존재하는 딕셔너리 타입

from collections import defaultdict

d1 = defaultdict(int)
d1["a"]             # int일 땐, 기본값이 0
print(d1)

d2 = defaultdict(list)   # list일 땐, 기본값이 빈 리스트 []
print(d2[0])

```

```python
# defaultdict를 이용한 카운트 정렬
def count_sort_dict(seq):
    
    result = []
    dic = defaultdict(list)
    
    # c의 key값(x)에 x를 append -> 각 key에는 x의 갯수만큼 x값이 append
    for num in seq:
        dic[num].append(num)
    
    # c의 key값의 최소값부터 최대값까지 extend 반복
    # -> 각 key에는 x값의 갯수만큼 x가 list형태로 저장되어 있으므로 extend
    for i in range(min(dic), max(dic) + 1):
        result.extend(dic[i])
    
    print(result)

seq = [3, 5, 2, 6, 8, 1, 0, 3, 6, 2, 5, 4, 1, 5, 3]
count_sort_dict(seq)
```

```python
# 병합 정렬
# 정렬 안되어 있는 리스트의 크기가 각각 1이 될때까지 리스트를 반으로 나눕니다.
# 리스트를 다시 정렬하면서 병합합니다.

def merge_sort(li):
    
    # 분할
    # 길이가 1인 리스트가 들어오면 return
    if len(li) < 2:
        return li
    
    # 길이가 1인 리스트로 쪼개질때까지 반복
    mid = len(li) // 2
    low = merge_sort(li[:mid])
    high = merge_sort(li[mid:])
    
    # 병합
    result = []           # 병합한 후 위 단계로 넘겨주기 위한 리스트
    l = h = 0
    
    while l < len(low) and h < len(high):
        if low[l] < high[h]:
            result.append(low[l])
            l += 1
            
        else:
            result.append(high[h])
            h += 1
        
    # 만약 low, high중 append하지 않은 요소 존재할 경우 처리
    result += low[l:]        # result.extend(low[l:])
    result += high[h:]       # result.extend(high[h:])
        
    return result
    
print(merge_sort([4,3,6,7,3,2,6,8,8,2,3,5]))
```

```python
# 병합정렬 최적화

# 리스트 병합
def merge(seq, left, mid, right):
    #print("left : {}\nright : {}".format(left, right))
    temp = [0] * len(seq)     # 정렬결과를 복제할 리스트
    i = left
    j = mid + 1
    k = left                  # temp리스트에서 가장 왼쪽부터 인덱스 이동
    
    # 작은 순서대로 temp 배열에 삽입
    while i <= mid and j <= right:
        if seq[i] < seq[j]:
            temp[k] = seq[i]
            i += 1
        else:
            temp[k] = seq[j]
            j += 1
        k += 1
        
    # i가 안끝났을 경우
    while i <= mid:
        temp[k] = seq[i]
        i += 1
        k += 1
        
    # j가 안끝났을 경우
    while j <= right:
        temp[k] = seq[j]
        j += 1
        k += 1
    
    # 정렬된 temp를 원래 리스트로 복제
    for t in range(left, right + 1):
        seq[t] = temp[t]
        
    print(seq)

# 리스트 분할
def merge_sort(seq, left, right):
    # 더이상 쪼개지지 않을 경우
    if left >= right:
        return
    
    mid = (left + right) // 2
    merge_sort(seq, left, mid)
    merge_sort(seq, mid + 1, right)
    merge(seq, left, mid, right)          # 리스트 병합
    
if __name__ == "__main__":
    seq = [5, 9, 7, 1, 3, 8, 2, 4]
    merge_sort(seq, 0, len(seq) - 1)
```

```python
# 퀵 정렬
# 1. 일반적으로 pivot을 첫번째 원소로 정한다
# 2. 나머지 리스트부터 left는 올라가며 pivot보다 작은값을, right은 내려가며 pivot보다 큰 값을 스캔한다.
# 3. 만일 left에서 pivot보다 큰값을, right에서 pivot보다 작은 값을 찾으면 서로 위치를 바꿔준다.
# 4. 다만, 위에서 left와 right가 엇갈린 경우엔 스캔이 완료 되었고 그 자리가 pivot의 자리가 되므로 바꿔준다.
# 5. 자리를 찾은 pivot을 기준으로 왼쪽, 오른쪽 리스트를 반복적으로 퀵정렬을 수행한다.

def quick_sort(seq, start, end):
    if start >= end:     # 원소가 1개인 경우 return
        return
    pivot = start        # 피벗은 첫번째 원소
    left, right = start + 1, end
    
    # left와 right가 엇갈린 경우 반복 종료
    while left <= right:
        while left <= end and seq[left] <= seq[pivot]:    # 범위 내에서 left가 피벗보다 클 경우 찾기
            left += 1
        while right > start and seq[right] >= seq[pivot]: # 범위 내에서 right가 피벗보다 작을 경우 찾기
            right -= 1
        # 엇갈린 경우 > right와 피벗 위치 바꿔주기, 피벗의 위치가 정해진다. 반복 종료
        if left > right:
            seq[pivot], seq[right] = seq[right], seq[pivot]
        # 엇갈리지 않았다면 left와 right 위치를 바꿔주기
        else:
            seq[left], seq[right] = seq[right], seq[left]
    
    print(seq)
    quick_sort(seq, start, right - 1)    # right는 자리를 찾은 pivot -> pivot - 1, pivot + 1
    quick_sort(seq, right + 1, end)
    
if __name__ == "__main__":
    seq = [5, 7, 9, 5, 0, 3, 1, 6, 2, 4, 8]
    quick_sort(seq, 0, len(seq) - 1)
    print(seq)
```

```python
# 최대 힙
class Heapify(object):
    def __init__(self, data = None):
        self.data = data or []
        
        for i in range(len(self.data) // 2, -1, -1):
            self.__max_heapify__(i)
            
    def __repr__(self):
        return repr(self.data)

    def parent(self, i):
        if i%2 == 0:
            return (i - 1) // 2
        else:
            return i // 2
        
    def left_child(self, i):
        return (2 * i) + 1
    
    def right_child(self, i):
        return (2 * i) + 2
    
    def __max_heapify__(self, i):
        largest = i
        left = self.left_child(i)
        right = self.right_child(i)
        n = len(self.data)
        
        if left < n and left and self.data[largest] < self.data[left]:
            largest = left
        if right < n and right and self.data[largest] < self.data[right]:
            largest = right
            
        if i != largest:
            self.data[i], self.data[largest] = self.data[largest], self.data[i]
            self.__max_heapify__(largest)
            
    def extract_max(self):
        
        if not self.data:
            return
        
        n = len(self.data)
        temp = self.data[0]
        self.data[0] = self.data[n - 1]
        self.data = self.data[:n - 1]
        self.__max_heapify__(0)
        
        return temp
    
    def insert(self, value):
        self.data.append(value)
        i = len(self.data) - 1
        
        while i > 0 and self.data[i] > self.data[self.parent(i)]:
            self.data[i] = self.data[self.parent(i)]
            i = self.parent(i)
            
        self.data[i] = value
    
            
if __name__ == "__main__":
    #seq = [3,5,2,6,8,1,0]
    #heap = Heapify(seq)
    heap = Heapify()
    for i in range(8, 0, -1):
        heap.insert(i)
    print(heap)
    
    while heap.data:
        print(heap.extract_max())
```

```python
# 힙 정렬
# 위에서 구현한 최대힙을 이용해서 힙 정렬
def heap_sort(seq):
    heap = Heapify(seq)
    result = []
    
    while heap.data:
        result.insert(0, heap.extract_max())
        
    print(result)

seq = [3,5,2,6,8,1,0]
heap_sort(seq)
```

```python
# 모듈을 이용해서 힙 정렬
import heapq

def heap_module_sort(seq):
    h = heapq.heapify(seq)      # seq 리스트를 heapify 수행

    result = []                 # heapify된 리스트의 요소를 정렬순으로 담아줄 리스트
    
    for i in range(len(seq)):
        result.append(heapq.heappop(seq))   # 순서대로 pop하여 append

    return result

seq = [3,5,2,6,8,1,0]
print(heap_module_sort(seq))
```

```python
# 버블 정렬        
def bubble_sort(li):
    length = len(li)
    
    for i in range(length - 1):
        for j in range(length - i - 1):
            if li[j] >= li[j + 1]:
                li[j], li[j + 1] = li[j + 1], li[j]
                
        print(li)

# 선택 정렬
def selection_sort(li):
    length = len(li)
    
    for i in range(length - 1):
        min_i = i
        
        for j in range(i + 1, length):
            if li[j] < li[min_i]:
                min_i = j
        li[min_i], li[i] = li[i], li[min_i]
        print(li)

# 삽입 정렬
def insertion_sort(li):
    length = len(li)
    
    for i in range(1, length):
        for j in range(i, 0, -1):
            if li[j] < li[j - 1]:
                li[j], li[j - 1] = li[j - 1], li[j]
        
        print(li)

# 놈 정렬
def gnome_sort(li):
    i = 0
    
    while i < len(li):
        if i == 0 or li[i - 1] <= li[i]:
            i += 1
            
        else:
            li[i - 1], li[i] = li[i], li[i - 1]
            i -= 1
            
    print(li)

# 카운트 정렬
def count_sort(li):
    from collections import defaultdict
    
    dic = defaultdict(list)
    result = []
    
    for l in li:
        dic[l].append(l)
        
    for i in range(min(dic), max(dic) + 1):
        result.extend(dic[i])
        
    print(result)
    
# 병합 정렬
# 리스트 분해
def merge_sort(seq, left, right):
    if left >= right:
        return
    
    mid = (left + right) // 2
    merge_sort(seq, left, mid)
    merge_sort(seq, mid + 1, right)
    merge(seq, left, mid, right)
# 리스트 병합    
def merge(seq, left, mid, right):
    temp = [0] * len(seq)
    i = left
    j = mid + 1
    k = left
    
    while i <= mid and j <= right:
        if seq[i] < seq[j]:
            temp[k] = seq[i]
            i += 1
        else:
            temp[k] = seq[j]
            j += 1
        k += 1
    
    while i <= mid:
        temp[k] = seq[i]
        i += 1
        k += 1
        
    while j <= right:
        temp[k] = seq[j]
        j += 1
        k += 1
        
    for t in range(left, right + 1):
        seq[t] = temp[t]
        
    #print(seq)
    
# 퀵 정렬
def quick_sort(seq, start, end):
    if start >= end:
        return
    
    pivot = start
    left = start + 1
    right = end
    
    while left <= right:
        while left <= end and seq[left] <= seq[pivot]:
            left += 1
        while right > start and seq[right] >= seq[pivot]:
            right -= 1
            
        if left > right:
            seq[pivot], seq[right] = seq[right], seq[pivot] 
        else:
            seq[left], seq[right] = seq[right], seq[left]
    
    print(seq)
    quick_sort(seq, start, right - 1)
    quick_sort(seq, right + 1, end)
    
if __name__ == "__main__":
    seq = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]
    quick_sort(seq, 0, len(seq) - 1)
    print(seq)
```

```python
def sqr_of_n(num):
    start = 0
    end = num
    
    while start <= end:
        mid = (start + end) // 2
        if mid ** 2 == num:
            return mid
        elif mid ** 2 < num:
            start = mid + 1
        else:
            end = mid - 1
    return end

num = int(input())
print(sqr_of_n(num))
```
