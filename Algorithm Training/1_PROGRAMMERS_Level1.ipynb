```python
# 정렬 (Level 1)
# K번째 수

def solution(array, commands):
    answer = []
    
    for i in range(len(commands)):
        start = commands[i][0] - 1   # 배열을 자를 시작점
        end = commands[i][1]         # 배열을 자를 끝점
        target = commands[i][2] - 1  # 자른 배열의 목표 인덱스
        
        # start~end까지 자른 배열을 정렬 후 target번째 원소를 append
        answer.append(sorted(array[start:end])[target])
    
    return answer
```

```python
# 완전탐색 (Level 1)
# 모의고사

def solution(answers):
    answer = []
    scores = [0] * 3        # 1 ~ 3번 수포자의 점수 배열
    
    first = [n for n in range(1, 6)]         # 1번 수포자의 찍는 방식
    second = [2, 1, 2, 3, 2, 4, 2, 5]        # 2번 수포자의 찍는 방식
    third = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]   # 3번 수포자의 찍는 방식
    
    for i in range(len(answers)):
        if answers[i] == first[i%len(first)]:     # %연산으로 모든 문제의 답을 1~3번 수포자의 답과 비교
            scores[0] += 1
        if answers[i] == second[i%len(second)]:
            scores[1] += 1
        if answers[i] == third[i%len(third)]:
            scores[2] += 1
    
    # 가장 높은 점수가 여럿일 경우 오름차순으로 반환해야하므로
    # 1 ~ 3번 수포자 순으로 최댓값과 비교해서 append
    for i in range(len(scores)):
        if scores[i] == max(scores):
            answer.append(i + 1)
    
    return answer
```

```python
# 해시 (Level 1)
# 완주하지 못한 선수
def solution(participant, complement):
    answer = ''
    
    # 단 한명만 완주하지 못했으므로, participant와 complement를 정렬하여 비교한다면 값(이름)이 다른 부분이 존재
    p = sorted(participant)
    c = sorted(complement)
    
    for i in range(len(c)):
        if c[i] != p[i]:
            answer = p[i]
            return answer
    # 예외 : p의 가장 맨뒤 주자가 완주 못했을 경우 > for문에서 검색 x
    answer = p[-1]
    
    return answer

participant = ["mislav", "stanko", "mislav", "ana"]
completion = ["stanko", "ana", "mislav"]

print(solution(participant, completion))
```

```python
# 로또의 최고 순위와 최저 순위 (Level1)

# 0은 알아볼 수 없는 숫자이므로
# 0을 모두 당첨 번호로 처리한다면 최고 순위
# 0을 모두 낙첨 번호로 처리한다면 최저 순위가 된다.

def solution(lottos, win_nums):
    answer = []
    
    # 2개 맞추면 7-2 > 5등, 6개 모두 맞추면 7-6 > 1등 이므로 최저순위를 일단 7로 놓는다.
    min_rank = 7
    zeros = 0     # 0의 갯수
    
    for num in lottos:
        if num in win_nums:
            min_rank -= 1
        if num == 0:
            zeros += 1
    # 최고 순위
    if min_rank - zeros == 7:    # 예외 : 0이 존재하지 않고 남은 숫자들 모두 당첨번호가 아닌 경우
        answer.append(min_rank - 1)   # 6등
    else:
        answer.append(min_rank - zeros)
    
    # 최저 순위
    if min_rank == 7:     # 예외 : 0을 제외한 번호가 모두 당첨번호가 아닌 경우
        answer.append(min_rank - 1)   # 6등
    else:
        answer.append(min_rank)
    
    return answer

l = [44, 1, 0, 0, 31, 25]
w = [31, 10, 45, 1, 6, 19]
solution(l, w)
```

```python
# 신규 아이디 추천(Level1)

# 1단계 : 대문자를 소문자로 치환
def stage1(new_id):
    return new_id.lower()

# 2단계 : 알파벳 소문자, 숫자, 빼기(-), 밑줄(_), 마침표(.) 제외한 문자 제거
def stage2(new_id):
    rule = "abcdefghijklmnopqrstuvwxyz1234567890-_."
    for c in new_id:
        if c not in rule:     # c가 rule안에 없다면, 빈 문자로 치환
            new_id = new_id.replace(c, "")

    return new_id

# 3단계 : 마침표(.)가 2번 이상 연속된 부분을 하나의 마침표로 치환
def stage3(new_id):
    while new_id.find("..") != -1:
        new_id = new_id.replace("..", ".")
    
    return new_id

# 4단계 : 처음이나 끝에 마침표(.) 존재하면 제거
def stage4(new_id):
    if len(new_id) > 1:
        if new_id[0] == '.':
            new_id = new_id[1:]
        if new_id[-1] == '.':
            new_id = new_id[:-1]
            
    else:               # 예외 : new_id가 "."만 남았을 경우, 인덱스 에러 나므로 빈문자 처리
        if new_id == '.':
            new_id = ""
        
    return new_id

# 5단계 : 빈 문자열이라면 "a"를 대입
def stage5(new_id):
    if new_id == "":
        new_id += "a"
    
    return new_id

# 6단계 : 첫 15개 문자 제외한 나머지 문자들 제거(길이 제한 15)
def stage6(new_id):
    if len(new_id) > 15:
        new_id = new_id[:15]
        
        if new_id[-1] == ".":         # 자르고 난 후에 맨뒤에 마침표가 존재하면 제거
            new_id = new_id[:-1]
    
    return new_id

# 7단계 : 길이가 2이하면 3이 될 때까지 마지막 문자를 반복해서 붙인다.
def stage7(new_id):
    while len(new_id) <= 2:
        new_id += new_id[-1]
        
    return new_id

def solution(new_id):
    answer = ''
    new_id = stage1(new_id)
    new_id = stage2(new_id)
    new_id = stage3(new_id)
    new_id = stage4(new_id)
    new_id = stage5(new_id)
    new_id = stage6(new_id)
    new_id = stage7(new_id)
    
    answer = new_id
    
    return answer
```

```python
# 포켓몬 (Level1)
# nums안의 포켓몬 종류의 수가 가져갈수 있는 포켓몬의 수(N/2)보다 적으면 종류의 수,
# 그렇지 않으면 가져갈 수 있는 포켓몬의 수를 반환하면 된다.

def solution(nums):
    from collections import Counter
    answer = 0
    c = Counter(nums)      # nums의 종류를 세기 위해 Counter 사용
    
    if len(nums) // 2 <= len(c):
        answer = len(nums) // 2
    else:
        answer = len(c)
    
    return answer
```
