```python
# 베스트 앨범
# Level 3

def solution(genres, plays):
    from collections import defaultdict
    answer = []
    
    dic = defaultdict(int)    # (장르 : 재생 횟수) 쌍의 디폴트 딕셔너리 생성
    for i in range(len(genres)):
        dic[genres[i]] += plays[i]   # 장르별로 재생 횟수를 누적

    # 재생 횟수기준으로 내림차순 정렬, sorted하고 다시 딕셔너리화 한 후 key값(장르)만 저장해준다.
    dic = dict(sorted(dic.items(), key = lambda x : x[1], reverse = True)).keys()

    for genre in dic:
        temp = {}              # (고유번호 : 재생 횟수) 쌍의 딕셔너리 생성
        for i in range(len(genres)):
            if genre == genres[i]:
                temp[i] = plays[i]
        # value값(재생횟수) 기준 내림차순 정렬 후, 딕셔너리화 해서 key값만 가져온다.
        # 2곡 까지 수록하고, 1곡일 시엔 1곡만 수록해야하므로 [:2]슬라이싱 후 결과에 extend해주기 위해 list화 해준다.
        temp = list(dict(sorted(temp.items(), key = lambda x : x[1], reverse = True)).keys())
        answer.extend(temp[:2])
    
    return answer
```

```python
# 디스크 컨트롤러
# 힙
# 작업 요청이 들어왔을 때 대기중인 작업중 가장 짧은 작업부터 수행

def solution(jobs):
    import heapq
    answer = 0
    count, now = 0, 0.   # count : 수행 작업 수, now : 현재 시간
    start = -1           # 작업 시작 시간
    waitings = []
    
    while count < len(jobs):
        for job in jobs:
            if start < job[0] <= now:
                heapq.heappush(waitings, [job[1], job[0]])  # 요청시간, 작업시간 swap해서 push
                                                        # 대기목록에서 작업시간이 적은 것부터 정렬하기 위해
        # 요청된 작업이 대기목록에 존재할 경우
        if len(waitings) > 0:
            temp = heapq.heappop(waitings)
            count += 1          # 작업이 수행됬으므로 수행 갯수 증가
            start = now         # 이번 작업의 시작시간을 start에 대입
            now += temp[0]      # 수행된 시간을 now(현재 시간)에 누적
            answer += now - temp[1]   # 결과에 현재 시간과 작업이 요청됬던 시간의 차를 더해줌
            
        else:
            now += 1
    answer = answer // len(jobs)
    return answer
```

```python
# 이중우선순위큐
# 힙
# Level3

def solution(operations):
    import heapq
    answer = []
    queue = []          # 이중우선순위큐
    inputs = []         # 입출력 명령을 담을 리스트

    for o in operations:
        inputs = o.split()
    
        if inputs[0] == "I":
            # (-값, 값)으로 push하여 최대값이 우선순위가 높은 heap으로 생성
            heapq.heappush(queue, (-1 * int(inputs[1]), int(inputs[1])))
    
        if inputs[0] == "D" and len(queue) > 0:   # queue가 비어있다면 D연산 무시
            if inputs[1] == "1":
                heapq.heappop(queue)       # 최대값을 제거(-값 기준으로 제일 작은 원소가 제거됨)

            if inputs[1] == "-1":          # 최솟값 제거
                # 임시 리스트에 queue를 원소의 2번째 값기준으로 정렬하여 생성
                temp = sorted(queue, key = lambda x : x[1])
                temp.pop(0)          # 원래값 기준으로 정렬되어있으므로 가장 앞의 원소 제거 > 최솟값 제거
                heapq.heapify(temp)  # 남은 원소들을 힙정렬
                queue = temp         # 다시 queue에 대입

    if len(queue) == 0:       # queue가 비어있다면 [0,0] 반환
        return [0, 0]
    
    else:
        answer.append(heapq.heappop(queue)[1])       # 최댓값 대입
        temp = sorted(queue, key = lambda x : x[1])  # queue를 다시한번 원래값 기준으로 정렬
        answer.append(temp[0][1])                    # 정렬 후 가장 맨앞 원소 대입(최솟값)
    return answer
```
