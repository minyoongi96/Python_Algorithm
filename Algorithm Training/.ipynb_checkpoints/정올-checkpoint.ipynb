```python
# 정올 Intermediate_Coder 이진탐색 Tutorial
def binarySearchRec(seq, target, low, high):
    if low > high:
        return -1
    
    mid = (low + high) // 2
    
    if target == seq[mid]:
        return mid
    elif target > seq[mid]:
        return binarySearchRec(seq, target, mid + 1, high)
    else:
        return binarySearchRec(seq, target, low, mid - 1)
    
    
# 리스트의 길이
n = int(input())
# 리스트
A = list(map(int, input().split(" ")))
# 찾는 숫자의 개수
targetN = int(input())
# 찾는 숫자 리스트
targets = list(map(int, input().split(" ")))
    
for i in targets:
    print(binarySearchRec(A, i, 0, len(A)), end = " ")
            
```

```python
# 제곱근의 정수부분 출력하기
def binary_search(low, high, target):
    while low <= high:
        mid = (low + high) // 2
        if mid ** 2 == target:
            return mid
        elif mid ** 2 < target:
            low = mid + 1
        else:
            high = mid - 1
    return high

num = int(input())

print(binary_search(0,num,num))
```

```python
# 정올-Beginner_Coder 자료처리 1번
# 1. "i a"는 a라는 수를 스택에 넣는다. 이때, a는 10,000 이하의 자연수이다.
# 2. "o"는 스택에서 데이터를 빼고, 그 데이터를 출력한다. 만약 스택이 비어있으면, "empty"를 출력한다.
# 3. "c"는 스택에 쌓여있는 데이터의 수를 출력한다.

class Stack(object):
    def __init__(self):
        self.stack = []
    
    def _push(self, value):
        self.stack.append(value)
            
    def _pop(self):
        if len(self.stack) == 0:
            print("empty")
        else:
            print(self.stack.pop())
    
if __name__ == "__main__":
    s = Stack()
    nFor = int(input())
    
    while nFor > 0:
        order = input().split(" ")
        
        if order[0] == "i" and len(order) > 1:
            s._push(int(order[1]))
            
        elif order[0] == "o":
            s._pop()
            
        elif order[0] == "c":
            print(len(s.stack))
            
        nFor -= 1
```

```python
# 정올-Beginner_Coder 자료처리 2번
# 큐
class Queue(object):
    def __init__(self):
        self.queue = []
        
    def enqueue(self, value):
        self.queue.append(value)
        
    def dequeue(self):
        if len(self.queue) == 0:
            print("empty")
        else:
            print(self.queue.pop(0))

def test():
    q = Queue()
    nFor = int(input())
    
    while nFor > 0:
        order = input().split(" ")
        
        if len(order) > 1:
            if order[0] == "i":
                num = int(order[1])
                q.enqueue(num)
            
        else:
            if order[0] == "o":
                q.dequeue()
            elif order[0] == "c":
                print(len(q.queue))
        
        nFor -= 1
        
if __name__ == "__main__":
    test()
```

```python
# 선택 정렬
def selection_sort(li, length):
    for i in range(length- 1):
        min_i = i
        
        for j in range(i,length):
            if li[j] < li[min_i]:
                min_i = j
                
        li[i] ,li[min_i] = li[min_i], li[i]
        
        for l in li:
            print(l, end=" ")
        print()
        
def testing():
    length = int(input())
    li = input().split(" ")
    
    for i in range(length):
        li[i] = int(li[i])
    
    selection_sort(li, length)
    
if __name__ == "__main__":
    testing()
```

```python
# 버블 정렬
def bubble_sort(li, length):
    for i in range(length - 1, 0, -1):
        for j in range(0,i):
            if li[j] > li[j+1]:
                li[j], li[j+1] = li[j+1], li[j]
        for l in li:
            print(l, end=" ")
        print()
        
def testing():
    length = int(input())
    li = input().split(" ")
    
    for i in range(length):
        li[i] = int(li[i])
    
    bubble_sort(li, length)
    
if __name__ == "__main__":
    testing()
```

```python
# 삽입 정렬
def insertion_sort(li, length):
    for i in range(1, length):
        for j in range(i, 0, -1):
            if li[j] < li[j - 1]:
                li[j], li[j - 1] = li[j - 1], li[j]
                
        for l in li:
            print(l, end = " ")
        print()
    
    
def testing():
    length = int(input())
    li = input().split(" ")
    
    for i in range(length):
        li[i] = int(li[i])
    
    insertion_sort(li, length)
        
testing()
```

```python
# 삽입정렬 횟수 세기
def insertion_sort(li, length):
    move = 0
    
    for i in range(1, length):
        for j in range(i, 0, -1):
            if li[j] < li[j - 1]:
                li[j], li[j - 1] = li[j - 1], li[j]
                move += 1
                
    print(move)
    
    
def testing():
    length = int(input())
    li = input().split(" ")
    
    for i in range(length):
        li[i] = int(li[i])
    
    insertion_sort(li, length)
        
testing()
```

```python
# 후위표기법
def testing(number_list):
    stack = []
    
    for value in number_list:
        if value == "+":
            A = stack.pop()
            B = stack.pop()
            stack.append(A + B)
            
        elif value == "-":
            A = stack.pop()
            B = stack.pop()
            stack.append(B - A)
            
        elif value == "*":
            A = stack.pop()
            B = stack.pop()
            stack.append(A * B)
            
        elif value == "/":
            A = stack.pop()
            B = stack.pop()
            stack.append(B // A)
            
        else:
            stack.append(int(value))

    print(stack.pop())
    
if __name__ == "__main__":
    N = int(input())
    number_list = list(input().split())
    
    testing(number_list)
```

```python
# 프로그래머스(해시)
# 완주하지 못한 선수
def solution(participant, completion):
    answer = ''
    
    p = sorted(participant)
    c = sorted(completion)
    
    for i in range(len(c)):
        if p[i] != c[i]:
            answer = p[i]
            return answer
        
    answer = p[i+1]
    
    return answer
```

```python
# 프로그래머스(힙)
# 더맵게
import heapq

def solution(scoville, K):
    answer = 0
    heap = []
    for s in scoville:
        heapq.heappush(heap, s)
    while heap[0] < K:
        try:
            temp = heapq.heappop(heap) + (heapq.heappop(heap) * 2)
        except IndexError:
            return -1
        heapq.heappush(heap, temp)
        answer += 1
    return answer
```
