```python
# 문자열 압축 (Level2)
def solution(s):
    answer = 0
    result = []   # 압축한 문자열들을 담아줄 리스트
    
    # s가 문자 한개일경우, 압축 불가능
    if len(s) == 1:
        return 1
    
    # i : 문자열 자르는 단위, 최대 범위는 s의 절반 까지
    for i in range(1, (len(s) // 2) + 1):
        cnt = 1   # 다음 문자열 단위와 같은지 갯수를 센다, 초기엔 1
        res = ''
        temp = s[:i]
        
        for j in range(i, len(s), i):
            if temp == s[j:j + i]:    # temp와 바로 다음 문자가 같다면 cnt +1
                cnt += 1
            else:
                if cnt != 1:          # cnt가 1보다 큰 경우, 기존에 중복해왔단 의미. cnt와 temp를 같이 대입
                    res += (str(cnt) + temp)
                else:                 # cnt가 1일 경우, 갯수가 한개이므로 temp를 그대로 대입
                    res += temp
                temp = s[j:j + i]     # temp를 다음 문자열로 이동
                cnt = 1               # cnt는 1로 다시 초기화
        # for문이 끝난 후 나머지 문자열들 처리
        if cnt != 1:
            res += (str(cnt) + temp)
        else:
            res += temp
            
        result.append(len(res))       # 길이를 리스트에 대입
    
    answer = min(result)              # 가장 짧은 길이를 반환
    
    return answer
```

```python
# 더 맵게
# 힙 (Level2)

def solution(scoville, K):
    import heapq
    answer = 0
    
    heapq.heapify(scoville)    # 힙 정렬 수행
    food1, food2 = 0, 0       # 스코빌 가장 낮은 음식, 두번째로 낮은 음식
    
    while scoville[0] <= K:    # 0번째 요소가 가장 스코빌 지수가 낮으므로
        if len(scoville) == 1: # K이상 스코빌 지수를 높일 수 없을때는 -1 반환
            return -1
        
        food1 = heapq.heappop(scoville)
        food2 = heapq.heappop(scoville)
        heapq.heappush(scoville, food1 + (food2 * 2))    # 음식을 섞고 다시 heappush
        # print(scovile)
        answer += 1           # 수행 횟수 +1
    
    return answer

scovile = [1, 2, 3, 9, 10, 12]
K = 7
solution(scovile, K)
```

```python
# 기능 개발
# 스택/큐 (Level 2)

def solution(progresses, speeds):
    answer = []
    
    while len(progresses) != 0:
        done = 0           # 오늘 완료한 작업의 수
        for i in range(len(progresses)):    # 작업 진도를 각각 더해준다.
            progresses[i] += speeds[i]
        
        # 작업 갯수가 0이 아니고 맨앞 작업진도가 100이상이면 > 조건 순서 주의!
        while len(progresses) != 0 and progresses[0] >= 100:
            progresses.pop(0)
            speeds.pop(0)
            done += 1
            
        if done != 0:
            answer.append(done)
        
    return answer

progresses = [93, 30, 55]
speeds = [1, 30, 5]
solution(progresses, speeds)
```

```python
# 다리를 지나는 트럭
# 스택/큐 (Level 2)
# 1초에 1만큼 트럭이 이동한다고 가정한다면
# 다리의 길이 만큼 [0]을 가지는 배열 bridge을 만들고 다리의 무게와 트럭의 무게를 비교하여
# bridge에 트럭을 append해주거나 빈 도로 0을 append하고 반복문에서 bridge의 맨 앞
# 요소를 pop하면서 경과시간을 +1씩 해줍니다. 

def solution(bridge_length, weight, truck_weights):
    answer = 0
    bridge = [0] * bridge_length      # 0은 빈 도로
    
    while len(bridge) != 0:
        answer += 1
        bridge.pop(0)
        
        if len(truck_weights) > 0:      # 대기중인 트럭이 남아있다면, 다리 무게를 계산해서 다음 트럭 대입
            if truck_weights[0] <= weight - sum(bridge):
                bridge.append(truck_weights.pop(0))
            else:         # 다리무게를 초과한다면 빈도로(0)를ㄴㄴ 추가
                bridge.append(0)
            
    
    return answer

bridge_length = 2
weight = 10
truck_weights = [7, 4, 5, 6]
print(solution(bridge_length, weight, truck_weights))
```

```python
# 프린터
# 스택/큐 (Level 2)

def solution(priorities, location):
    answer = 0
    
    # 인덱스 리스트 추가, priorities 리스트의 인덱스 정보
    index_list = [i for i in range(len(priorities))]
    
    while True:
        #print(priorities)
        maximum = max(priorities)
        if priorities[0] < maximum:   # 0번째 값이 최고 우선순위가 아닐시, 맨 뒤로 이동
            priorities.append(priorities.pop(0))
            index_list.append(index_list.pop(0))
        else:       # 최고 우선순위 일시, 프린터 완료, answer +1
            answer += 1
            if index_list[0] == location:    # location번째의 작업일 경우 return
                return answer
            else:  # location번째의 작업이 아닐경우 필요 없으니 pop하고 다음 작업 실행
                priorities.pop(0)
                index_list.pop(0)


priorities = [2, 1, 3, 2]
location = 2
solution(priorities, location)
```

```python
# 소수 찾기

from itertools import permutations

def check(num):
    if num < 2:
        return False
    
    for i in range(2, num):
        if num % i == 0:
            return False
        
    return True

def solution(numbers):
    answer = 0
    nums = [n for n in numbers]
    per = []     # 순열 조합을 담아둘 리스트
    result = []  # 찾은 소수들을 담아줄 리스트
    
    for i in range(1, len(numbers) + 1):
        # 1부터 numbers의 길이만큼 순열 조합, join을 이용하여 병합
        per += list(map(''.join, permutations(nums, i)))
        
    for i in list(set(per)):   # set을 이용하여 순열조합의 중복값 처리
        if check(int(i)):      # 소수인지 확인
            result.append(int(i))  # check함수의 반환값이 True면 result에 append
    
    #  위에 소수인지 확인할때 int형으로 값을 변환할때, "011" > 11, "11" > 11로 되므로 중복값 발생 가능
    answer = len(set(result))      # 중복 제거
        
    return answer

numbers = "011"
solution(numbers)
```

```python
# 가장 큰수

# numbers안의 숫자문자열을 내림차순으로 정렬해서 하나의 숫자로 join
# 문자열은 정렬할 때 ASCII코드 기준으로 정렬하므로
# 문제에서 0이상 1000이하 라고 제한이 있기 때문에 모든 수가 세자리 이상이 되도록
# *3을 한 후 내림차순으로 정렬하고 join하면 끝
# join한 후에 int로 변환, 다시 str로 변환하여 반환한다. (0이 잘못 처리될수 있는 경우 처리)

def solution(numbers):
    answer = ''
    
    numbers = list(map(str, numbers))     # numbers의 요소들을 모두 str형으로 변환
    numbers.sort(key = lambda x : x * 3, reverse = True)   # 모든 원소에 *3한 후 내림차순으로 정렬
    
    answer = str(int(''.join(numbers)))   # 하나의 문자로 join
    
    return answer
```

```python
# 메뉴 리뉴얼

from itertools import combinations
from collections import Counter

def solution(orders, course):
    answer = []
    
    for i in course:      # combinations를 통해 i개만큼 메뉴의 조합들을 구함
        c = Counter()     # 단품메뉴들의 수를 세기 위해 Counter() 선언
        
        for order in orders:
            temp = list(order)   # "ABC"를 ["A", "B", "C"]로 변환
            
            # 단품메뉴들을 i개씩 조합을 만들어준다.
            # temp를 정렬하여 "AB"와 "BA"로 결과를 따로 만들어주지 않기 위해
            c.update(list(map(''.join, combinations(sorted(temp), i))))
        
        # c.most_common()[0][1]
        # c.most_common() : c라는 카운터에서 최빈값을 가지는 데이터부터 (key, value)쌍의 튜플 형식으로 정렬한다
        # 정렬된 부분 중 [0][1]이므로 가장 앞에있는 즉, 최빈값을 가지는 튜플중 [1]번째 값 -> 나온 수가 된다.
        if len(c) > 0:    # 카운터가 존재할 경우에만 실행
            high = c.most_common()[0][1]
        
            for key, value in c.items():      # key: i개의 주문조합, value: 빈도 수
                if value == high and value > 1:    # 두사람 이상이 시킨 주문 중 가장 빈도가 높은 주문
                    answer.append(key)
    
    answer = sorted(answer)
    return answer
orders = ["XYZ", "XWY", "WXA"]
course = [2, 3, 4]
solution(orders, course)
```

```python
# 전화번호 목록
# 나의 풀이
def solution(phone_book):
    answer = True
    phone_book = sorted(phone_book)     # 전화번호 목록을 정렬
                                        #  > 접두어 관계라면 바로 뒤에 위치할 것
    
    for i in range(len(phone_book) - 1):
        if len(phone_book[i]) < len(phone_book[i + 1]):    # 접두어가 되는 [i]가 [i+1]보다 짧아야함
            if phone_book[i] == phone_book[i + 1][:len(phone_book[i])]:
                answer = False
                return answer
            
    return answer
```

```python
# 주식 가격
# 스택/큐

def solution(prices):
    from collections import deque
    
    answer = []
    queue_prices = deque(prices)
    
    # queue를 활용하여 가격이 떨어지기 전까지 시간을 구함
    while queue_prices:
        price = queue_prices.popleft()
        time = 0
        
        for p in queue_prices:
            time += 1
            if price > p:           # 가격이 떨어졌다면 반복문 break
                break
        answer.append(time)
        
    return answer
```

```python
# H-Index
# 정렬
# Level2

# 논문을 오름차순으로 정렬하고 인덱스와 인용횟수(인덱스에 해당하는 리스트 원소)를 비교합니다.
# 인용횟수는 0 이상 10000회 이하 이므로 인덱스 번호를 0부터 부여하므로 인덱스 >= 인용횟수 가 될때
# 인덱스의 값을 H-Index로 반환합니다.

def solution(citations):
    citations = sorted(citations, reverse = True)   # 오름차순 정렬
    answer = 0
    
    for i, c in enumerate(citations):     # 인덱스 0부터 리스트의 원소와 비교
        if i >= c:
            return i
    
    return len(citations)
```

```python
# 위장
# 해시
# 문제에서 해석해보면, 같은 종류의 옷은 겹치지 않아야 하고
# 적어도 하나의 옷은 입어야하는 경우의 수를 구하면 될거 같습니다.
# 옷 종류별로 나눈 후 종류의 수 + 1(+1은 안입은 경우를 포함)를 모두 곱하고
# 아무것도 안입은 경우는 없으므로 -1을 해서 반환해줍니다.

def solution(clothes):
    answer = 1
    c = []
    
    clothes = sorted(clothes, key = lambda x : x[1])
    c.append([clothes[0][0]])
    category = clothes[0][1]
    idx = 0
    
    for i in range(1, len(clothes)):
        if clothes[i][1] == category:
            c[idx].append(clothes[i][0])
        else:
            c.append([clothes[i][0]])
            idx += 1
            category = clothes[i][1]
    
    for cloth in c:
        answer *= len(cloth) + 1
    
    return answer - 1
clothes = [["yellowhat", "headgear"], ["bluesunglasses", "eyewear"],
           ["green_turban", "headgear"]]
solution(clothes)
```

```python
# 타겟 넘버 - Level 2
# DFS, BFS

# DFS 풀이
answer = 0

def dfs(start, numbers, target, result):    # solution 메소드에서 재귀 호출을 위해 dfs 메소드 따로 정의
    global answer   # 전역 변수로 선언
    
    if start == len(numbers):
        if result == target:
            answer += 1
        return    # 모두 탐색했으므로 return
    
    # -, + 를 나누어 재귀 호출
    dfs(start + 1, numbers, target, result + numbers[start])
    dfs(start + 1, numbers, target, result - numbers[start])

def solution(numbers, target):
    global answer
    
    dfs(0, numbers, target, 0)    # 맨 앞 0부터 dfs 실행
    
    return answer
```

```python
# BFS 풀이

from collections import deque

def solution(numbers, target):
    answer = 0
    queue = deque()
    queue.append((0, 0))     # (현재 확인중인 인덱스, 결과값) 삽입
    
    while queue:
        idx, result = queue.popleft()
        
        if idx == len(numbers):
            if result == target:
                answer += 1
        else:
            # 다음 확인할 인덱스와 현재 숫자를 누적합에 +, - 해서 삽입
            queue.append(idx + 1, result + numbers[idx])
            queue.append(idx + 1, result - numbers[idx])
```

```python
# 네트워크 - Level 3
# DFS, BFS

# BFS 풀이
from collections import deque

def solution(n, computers):
    answer = 0
    visited = [False] * n
    queue = deque()
    
    for v in range(n):
        if visited[v] is False:    # 방문한 적이 없다면
            queue.append(v)        # 해당 컴퓨터 번호 입력
            
            while queue:   # 큐가 비어질때까지
                current = queue.popleft()   # 현재 확인중인 컴퓨터의 번호
                visited[current] = True     # 방문 처리
                
                # 자기 자신이 아니고, 다른 컴퓨터와 관계가 1일 경우
                for c in range(len(computers[current])):
                    if c != current and computers[current][c] == 1:
                        if visited[c] is False:       # 방문하지 않은 컴퓨터라면
                            queue.append(c)
            answer += 1
        # 방문한 적이 있다면
        else:
            continue
            
    return answer
```

```python
# DFS 풀이

def dfs(current, n, computers, visited):
    visited[current] = True    # 현재 컴퓨터 방문 처리
    
    for i in range(n):
        if i != current and computers[current][i] == 1:   # 연결된 컴퓨터라면
            if visited[i] is False:   # 아직 방문하지 않은 컴퓨터라면
                dfs(i, n, computers, visited)

def solution(n, computers):
    answer = 0
    
    visited = [False] * n
    for current in range(n):
        if visited[current] is False:
            dfs(current, n, computers, visited)
            answer += 1     # dfs를 모두 수행하면 연결
    
    return answer
```
